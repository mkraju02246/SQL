## DDL - Data  Defination Language

  ### All those operations which can change the data structure of the table come under DDL

## DML - Data Manipulation Language
  ### All those operations which can manipulate the data a table are called as DML

## DCL - Data Control Language
  ### All those operations which control the access of data (authority to access data) come under DCL

## TCL - Transaction Control Language
  ### All those operations which control the effect of transaction come under TCL
